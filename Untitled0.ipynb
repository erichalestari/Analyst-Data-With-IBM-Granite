```python
!pip install langchain-community
!pip install replicate
from langchain_community.llms import Replicate
import os
from google.colab import userdata
api_token = userdata.get('api_token')
os.environ["REPLICATE_API_TOKEN"] = api_token
model = "ibm-granite/granite-3.3-8b-instruct"
output1 = Replicate(
    model=model,
    replicate_api_token = api_token
)
```

```python
import pandas as pd

diaster_data = pd.read_excel("mm.xlsx")
diaster_data.head()
```

```python
reviews_text = "\n".join([f"Review {i+1}: {review}"for i, review
                          in enumerate (diaster_data)])
prompt1 = f"""
from the diaster data, provide a brief report in plain english on:
focus on the following:
1. total number of diaster events.
2. the most frequency diaster types.
3. top 10 states with the most diaster events.
4. average diaster duration in days.
5. distribution of diaster duration (histogram).
6. relationship between diaster type and assistance program provided.
{reviews_text}
"""
Response = output1.invoke(prompt1)
print("Granite Model Response:\n")
print(Response)
```

```python
prompt2 = f"""
Based on this disaster data, please summarize the final results of disaster risk modeling.

Specifically, I want to see:
- The top 5 U.S. states with the highest risk of disasters.
- For each of those states, what is the most common disaster type?
- What is the probability (%) of that disaster type occurring in that state?
- A brief explanation of which regions are at greatest risk and why.

Show only the final analysis results in plain English and a summary table. Do not show code or describe how the analysis was done.
{reviews_text}
"""
Response = output1.invoke(prompt2)
print("Granite Model Response:\n")
print(Response)
```

```python
prompt3 = f"""
based on diaster data, please make expected loss and summarize the risk model of natural disasters,

In particular, I would like to see
- the expected loss of the 5 countries with the highest risk
- and estimate the amount of premium to be paid based on the expected loss of the 5 countries with the highest natural disaster risk.
{reviews_text}
"""
Response = output1.invoke(prompt3)
print("Granite Model Response:\n")
print(Response)
```
